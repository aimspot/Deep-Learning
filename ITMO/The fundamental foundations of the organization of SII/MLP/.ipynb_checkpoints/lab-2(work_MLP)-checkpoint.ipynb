```python
X = df.iloc[:, 1:]
Y = df.iloc[:, 0]
```

```python
train_percent = 80 # Процент тестовой формирует исходя из значения обучающей выборки
```

```python
def train_test_split_X(data, train_percent):
    train = list()
    test = list()
    for i in range(0, len(data)):
        if i <= len(data) * 0.8:
            train.append(data.iloc[i,:])
        else:
            test.append(data.iloc[i,:])
    return np.array(train), np.array(test)
```

```python
def train_test_split_Y(data, train_percent):
    train = list()
    test = list()
    for i in range(0, len(data)):
        if i <= len(data) * 0.8:
            train.append(data[i])
        else:
            test.append(data[i])
    return np.array(train), np.array(test)
```

```python
trainX, testX = train_test_split_X(X, train_percent)
trainY, testY = train_test_split_Y(Y, train_percent)
```

```python
class MLP:
    def __init__(self, arch, lr, epochs):
        self.arch = len(arch) - 1
        self.W = self.__init_w__(arch)
        self.b = self.__init_b__(arch)
        self.lr = lr
        self.epochs = epochs
#         print("W_0", self.W[0])
#         print("W_1", self.W[1])
        self.b = self.__init_b__(arch)
        #print("W_1", self.b)
        

    def __init_w__(self, arch):
        return  [rng.uniform(-1, 1, size=(arch[i+1], arch[i])) for i in range(self.arch)]
    
    def __init_b__(self, arch):
        return [rng.uniform(-1, 1, size=(1, arch[i+1])) for i in range(self.arch)]
    
    def FeedForward(self, w, x, b):
        f = np.dot(w,x)
        f = f + b
        return self.sigmoid(f)
    
    def BackPropagation(self, x, y, z, w, b):
        self.delta = []
        self.dW = []
        self.dB = []
#         print("z len(z)-1",z[len(z)-1])
#         print("Y",y)
        # Считаем ошибку выходного слоя
        self.delta.append((z[len(z)-1] - y) * self.Bsigmoid(z[len(z)-1]))
        #print("Delta", self.delta)
        
        for i in range(0, self.arch-1):
            self.delta.append(np.dot(self.delta[i], w[len(z) - 1 - i])*self.Bsigmoid(z[len(z)- 2 - i]))
        self.delta = np.flip(self.delta, 0)
        #self.delta = self.delta/np.array(x).shape[0] 
        
        #print("Delta", self.delta)
        
        self.dW.append(w[0] - self.lr * np.kron(self.delta[0], x).reshape( len(z[0]), np.array(x).shape[0] ) )
        self.dB.append(b[0] - self.lr * self.delta[0])
        for i in range(1, self.arch):
            self.dW.append( w[i] - self.lr * np.kron(self.delta[i], z[i-1]).reshape(len(z[i]), len(z[i-1])) )
            self.dB.append( b[i] - self.lr * self.delta[i] )
            
        
#         print("Self.dW", self.dW)
        return np.array(self.dW), self.dB
    
    def mse(self, y_pred, y):
        err = np.square(y_pred - y)
        return err
    
    def sigmoid(self, x):
        return 1/(1+ np.exp(- x))
    
    def Bsigmoid(self, x):
        return 1/(1+ np.exp(- x))*(1-1/(1+ np.exp(- x)))
    
    def fit(self, trainX, trainY):
        for k in range(0, self.epochs):
            #print("OLD W", self.W)
            for i in range(len(trainX)):
                self.z = list([])
                for j in range(self.arch):
                    #print("----------------------")
                    #print("Self.w", self.W[0])
                    if j == 0:
    #                     print("X_i", trainX[i])
    #                     print("W_j", self.W[j])
                        self.z.append(self.FeedForward(self.W[j], trainX[i], self.b[0][j]))
                    else:
                        self.z.append(self.FeedForward(self.W[j], self.z[j-1], self.b[j][0]))
                    #print("Z", self.z)
                #backprop
                self.W, self.b = self.BackPropagation(trainX[i], trainY[i], self.z, self.W, self.b)
                # loss
            #print("New W", self.W)
            answer = self.predict(trainX,trainY)
            mse_answer = list(map(self.mse, answer, trainY))
            print(f"Epoch:{k} ----  MSE {sum(mse_answer)/np.array(trainY).shape[0]}")
    
    def predict(self, trainX, trainY):
        self.predicts = []
        for i in range(0, np.array(trainX).shape[0]):
            self.z = []
            for j in range(self.arch):
                if j == 0:
                    self.z.append(self.FeedForward(self.W[j], trainX[i], self.b[0][j]))
                else:
                    self.z.append(self.FeedForward(self.W[j], self.z[j-1], self.b[j][0]))
            index, = np.where(self.z[-1] == (reduce(max, self.z[-1])))[0]
            self.predicts.append(index)
        #print(self.predicts)
        return(self.predicts)
            
        
            
```

```python
md = (22, 16, 8, 2)
model = MLP(md,lr = 0.001, epochs = 150)
model.fit(trainX, trainY)
```

```python
predictions = model.predict(testX, testY)
```

```python
 def metric(testY, pred):
        TP = 0
        TN = 0
        FP = 0
        FN = 0
        for i,j in zip(testY, pred):
            if i == 0 and j == 0:
                TP = TP + 1
            elif i == 1 and j == 0:
                FP = FP + 1
            elif i == 1 and j == 1:
                TN = TN + 1
            elif i==0 and j==1:
                FN = FN + 1
        return TP, TN, FP, FN
            
```

```python
True_Positive, True_Negative, False_Positive, False_Negative = metric(testY, predictions)
```

```python
accuracy = (True_Positive + True_Negative)/(True_Positive + True_Negative + False_Positive + False_Negative)
print(f"Accuracy: {accuracy}")
```

```python
recall = (True_Positive)/(True_Positive + False_Negative)
print(f"Recall: {recall}")
```

```python
precision = (True_Positive)/(True_Positive + False_Positive)
print(f"Precision: {precision}")
```

```python
f1 = 2* ((precision * recall) / (precision + recall))
print(f"F-1: {f1}")
```

```python
def plot_roc_curve(fper, tper):
    plt.plot(fper, tper, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic Curve')
    plt.legend()
    plt.show()
```

```python
fper, tper, thresholds = roc_curve(testY, predictions)
plot_roc_curve(fper, tper)
```
